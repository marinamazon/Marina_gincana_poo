## PARTE 2 ‚Äì M√©todos e Heran√ßa

**Tarefas:**
- Criar pelo menos 2 m√©todos de a√ß√£o para a classe principal.
- Criar uma subclasse herdando da classe principal.
- Adicionar pelo menos 1 novo atributo e 1 novo m√©todo exclusivo na subclasse.
- Utilizar `super()` no construtor da subclasse.

**C√≥digo:**

```python
# Desenvolva aqui o c√≥digo da Parte 2
# Classe principal
class Gincana:
    def __init__(self, nome, data):
        self.nome = nome
        self.data = data
        self.equipes = []

    def adicionar_equipe(self, equipe):
        self.equipes.append(equipe)
        print(f"Equipe '{equipe}' adicionada √† gincana '{self.nome}'.")

    def listar_equipes(self):
        print(f"Equipes da gincana '{self.nome}':")
        for equipe in self.equipes:
            print(f"- {equipe}")

# Subclasse herdando de Gincana
class GincanaTematica(Gincana):
    def __init__(self, nome, data, tema):
        super().__init__(nome, data)  # Utilizando super() para herdar
        self.tema = tema  # Novo atributo exclusivo

    def mostrar_tema(self):  # Novo m√©todo exclusivo da subclasse
        print(f"O tema da gincana '{self.nome}' √©: {self.tema}")

# Teste do c√≥digo
g1 = Gincana("Gincana de Matem√°tica", "2025-06-20")
g1.adicionar_equipe("Alfa")
g1.adicionar_equipe("Beta")
g1.listar_equipes()

print("\n---\n")

gt = GincanaTematica("Gincana Cultural", "2025-07-15", "Folclore Brasileiro")
gt.adicionar_equipe("Tup√£")
gt.adicionar_equipe("Curupira")
gt.listar_equipes()
gt.mostrar_tema()
```

## PARTE 3 ‚Äì Encapsulamento e Abstra√ß√£o

**Tarefas:**
- Tornar pelo menos 1 atributo da classe principal privado (`__atributo`).
- Criar m√©todos `get` e `set` para o atributo privado.
- Implementar um m√©todo que contenha l√≥gica interna (ex: c√°lculo, verifica√ß√£o).

**C√≥digo:**

```python
# Desenvolva aqui o c√≥digo da Parte 3
class Gincana:
    def __init__(self, nome, data):
        self.__nome = nome  # Atributo privado
        self.data = data
        self.equipes = {}

    # M√©todos get e set para o nome (encapsulamento)
    def get_nome(self):
        return self.__nome

    def set_nome(self, novo_nome):
        if len(novo_nome.strip()) >= 3:
            self.__nome = novo_nome
        else:
            print("Nome inv√°lido. Deve ter pelo menos 3 letras.")

    # Adicionar equipe com pontua√ß√£o inicial 0
    def adicionar_equipe(self, equipe):
        if equipe not in self.equipes:
            self.equipes[equipe] = 0
            print(f"Equipe '{equipe}' adicionada.")
        else:
            print("Equipe j√° cadastrada.")

    # Atribuir pontos a uma equipe (l√≥gica interna)
    def registrar_pontos(self, equipe, pontos):
        if equipe in self.equipes:
            self.equipes[equipe] += pontos
            print(f"{pontos} ponto(s) adicionados √† equipe '{equipe}'.")
        else:
            print("Equipe n√£o encontrada.")

    # Mostrar pontua√ß√£o
    def mostrar_pontuacoes(self):
        print(f"Pontua√ß√µes da gincana '{self.__nome}':")
        for equipe, pontos in self.equipes.items():
            print(f"- {equipe}: {pontos} ponto(s)")

# Teste
g = Gincana("Gincana de Ci√™ncias", "2025-09-10")
print("Nome original:", g.get_nome())

g.set_nome("Ci√™ncia Show")  # Setter com valida√ß√£o
print("Novo nome:", g.get_nome())

g.adicionar_equipe("Equipe Azul")
g.adicionar_equipe("Equipe Verde")

g.registrar_pontos("Equipe Azul", 10)
g.registrar_pontos("Equipe Verde", 15)

g.mostrar_pontuacoes()
```

## PARTE 4 ‚Äì Polimorfismo e Banco de Dados

**Tarefas:**
- Implementar polimorfismo: sobrescrever um m√©todo na subclasse.
- Integrar o sistema com banco de dados (MySQL ou SQLite).
- Implementar inser√ß√£o (INSERT), consulta (SELECT) e exclus√£o (DELETE) de registros no banco de dados.

**C√≥digo:**

```python
# Desenvolva aqui o c√≥digo da Parte 4
import sqlite3

# Classe base
class Gincana:
    def __init__(self, nome, data):
        self.nome = nome
        self.data = data

    def descricao(self):
        return f"Gincana: {self.nome} - Data: {self.data}"

    def salvar_no_banco(self):
        conn = sqlite3.connect("gincanas.db")
        cursor = conn.cursor()
        cursor.execute("""
            CREATE TABLE IF NOT EXISTS gincanas (
                id INTEGER PRIMARY KEY AUTOINCREMENT,
                nome TEXT,
                data TEXT,
                tipo TEXT
            )
        """)
        cursor.execute("INSERT INTO gincanas (nome, data, tipo) VALUES (?, ?, ?)",
                       (self.nome, self.data, "Comum"))
        conn.commit()
        conn.close()

    @staticmethod
    def listar_gincanas():
        conn = sqlite3.connect("gincanas.db")
        cursor = conn.cursor()
        cursor.execute("SELECT id, nome, data, tipo FROM gincanas")
        registros = cursor.fetchall()
        conn.close()

        print("üìã Gincanas cadastradas:")
        for r in registros:
            print(f"ID: {r[0]} | Nome: {r[1]} | Data: {r[2]} | Tipo: {r[3]}")

    @staticmethod
    def excluir_gincana(id_gincana):
        conn = sqlite3.connect("gincanas.db")
        cursor = conn.cursor()
        cursor.execute("DELETE FROM gincanas WHERE id = ?", (id_gincana,))
        conn.commit()
        conn.close()
        print(f"Gincana com ID {id_gincana} exclu√≠da.")

# Subclasse com polimorfismo
class GincanaTematica(Gincana):
    def __init__(self, nome, data, tema):
        super().__init__(nome, data)
        self.tema = tema

    def descricao(self):  # POLIMORFISMO (m√©todo sobrescrito)
        return f"Gincana Tem√°tica: {self.nome} - Data: {self.data} - Tema: {self.tema}"

    def salvar_no_banco(self):
        conn = sqlite3.connect("gincanas.db")
        cursor = conn.cursor()
        cursor.execute("""
            CREATE TABLE IF NOT EXISTS gincanas (
                id INTEGER PRIMARY KEY AUTOINCREMENT,
                nome TEXT,
                data TEXT,
                tipo TEXT
            )
        """)
        cursor.execute("INSERT INTO gincanas (nome, data, tipo) VALUES (?, ?, ?)",
                       (self.nome, self.data, f"Tem√°tica: {self.tema}"))
        conn.commit()
        conn.close()

# Testes
g1 = Gincana("Gincana Esportiva", "2025-09-01")
print(g1.descricao())
g1.salvar_no_banco()

g2 = GincanaTematica("Gincana Cultural", "2025-09-20", "Cultura Nordestina")
print(g2.descricao())  # Polimorfismo
g2.salvar_no_banco()

print("\n--- CONSULTA ---")
Gincana.listar_gincanas()

print("\n--- EXCLUS√ÉO ---")
Gincana.excluir_gincana(1)  # Exclui gincana com ID 1 (ajuste conforme seus dados)
Gincana.listar_gincanas()
```

```python
# Desenvolva aqui o c√≥digo da Parte 5]
import sqlite3

# === CLASSE PRINCIPAL: GINCANA ===
class Gincana:
    def __init__(self, nome, data):
        self.__nome = nome  # Atributo privado (ENCAPSULAMENTO)
        self.data = data
        self.equipes = {}

    # Get e Set do nome (Encapsulamento)
    def get_nome(self):
        return self.__nome

    def set_nome(self, novo_nome):
        if len(novo_nome.strip()) >= 3:
            self.__nome = novo_nome
        else:
            print("Nome inv√°lido.")

    # Adiciona uma equipe √† gincana
    def adicionar_equipe(self, equipe):
        if equipe not in self.equipes:
            self.equipes[equipe] = 0
            print(f"‚úÖ Equipe '{equipe}' adicionada.")
        else:
            print("‚ö†Ô∏è Equipe j√° existe.")

    # Atribui pontos a uma equipe (ABSTRA√á√ÉO com l√≥gica)
    def registrar_pontos(self, equipe, pontos):
        if equipe in self.equipes:
            self.equipes[equipe] += pontos
        else:
            print("‚ùå Equipe n√£o encontrada.")

    # Mostra as pontua√ß√µes
    def mostrar_pontuacoes(self):
        print(f"üìã Pontua√ß√£o da gincana '{self.__nome}':")
        for equipe, pontos in self.equipes.items():
            print(f"- {equipe}: {pontos} ponto(s)")

    # Descri√ß√£o da gincana (para polimorfismo)
    def descricao(self):
        return f"Gincana comum: {self.__nome} | Data: {self.data}"

    # SALVAR NO BANCO
    def salvar_no_banco(self):
        conn = sqlite3.connect("gincanas.db")
        cursor = conn.cursor()
        cursor.execute("""
            CREATE TABLE IF NOT EXISTS gincanas (
                id INTEGER PRIMARY KEY AUTOINCREMENT,
                nome TEXT,
                data TEXT,
                tipo TEXT
            )
        """)
        cursor.execute("INSERT INTO gincanas (nome, data, tipo) VALUES (?, ?, ?)",
                       (self.__nome, self.data, "Comum"))
        conn.commit()
        conn.close()

    # LISTAR DO BANCO
    @staticmethod
    def listar_gincanas():
        conn = sqlite3.connect("gincanas.db")
        cursor = conn.cursor()
        cursor.execute("SELECT id, nome, data, tipo FROM gincanas")
        registros = cursor.fetchall()
        conn.close()

        print("\nüìö Gincanas cadastradas:")
        for r in registros:
            print(f"ID: {r[0]} | Nome: {r[1]} | Data: {r[2]} | Tipo: {r[3]}")

    # EXCLUIR DO BANCO
    @staticmethod
    def excluir_gincana(id_gincana):
        conn = sqlite3.connect("gincanas.db")
        cursor = conn.cursor()
        cursor.execute("DELETE FROM gincanas WHERE id = ?", (id_gincana,))
        conn.commit()
        conn.close()
        print(f"üóëÔ∏è Gincana com ID {id_gincana} foi exclu√≠da.")

# === SUBCLASSE: GINCANA TEM√ÅTICA ===
class GincanaTematica(Gincana):
    def __init__(self, nome, data, tema):
        super().__init__(nome, data)  # super() = herda do construtor da classe pai
        self.tema = tema

    # POLIMORFISMO: sobrescrevendo o m√©todo descricao()
    def descricao(self):
        return f"Gincana Tem√°tica: {self.get_nome()} | Data: {self.data} | Tema: {self.tema}"

    # Sobrescrevendo salvar para incluir tema
    def salvar_no_banco(self):
        conn = sqlite3.connect("gincanas.db")
        cursor = conn.cursor()
        cursor.execute("""
            CREATE TABLE IF NOT EXISTS gincanas (
                id INTEGER PRIMARY KEY AUTOINCREMENT,
                nome TEXT,
                data TEXT,
                tipo TEXT
            )
        """)
        cursor.execute("INSERT INTO gincanas (nome, data, tipo) VALUES (?, ?, ?)",
                       (self.get_nome(), self.data, f"Tem√°tica: {self.tema}"))
        conn.commit()
        conn.close()

# === TESTES DO SISTEMA FINAL ===

print("=== TESTE 1: Gincana comum ===")
g1 = Gincana("Gincana Esportiva", "2025-10-01")
g1.adicionar_equipe("Le√µes")
g1.adicionar_equipe("Tigres")
g1.registrar_pontos("Le√µes", 10)
g1.registrar_pontos("Tigres", 20)
g1.mostrar_pontuacoes()
print(g1.descricao())
g1.salvar_no_banco()

print("\n=== TESTE 2: Gincana tem√°tica ===")
g2 = GincanaTematica("Gincana Cultural", "2025-11-05", "Cultura Afro-Brasileira")
g2.adicionar_equipe("Mandinga")
g2.registrar_pontos("Mandinga", 30)
g2.mostrar_pontuacoes()
print(g2.descricao())
g2.salvar_no_banco()

print("\n=== CONSULTA AO BANCO ===")
Gincana.listar_gincanas()

print("\n=== EXCLUS√ÉO DE GINCANA (ID 1) ===")
Gincana.excluir_gincana(1)

print("\n=== CONSULTA FINAL ===")
Gincana.listar_gincanas()
```
